# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urunEkle(ad,fiyat):
    with open("urunler.txt","w",encoding="utf-8") as file:
        file.write(f"{ad} isimli urunun fiyati = {fiyat}")
        





ad = input("Ad : ")
fiyat = float(input("Fiyat:" ))
urunEkle(ad,fiyat)

````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def kelimeDegistir(dosyaAdi,eskiKelime,yeniKelime):

    with open(dosyaAdi,"r",encoding="utf-8") as file:
        icerik = file.read()
    yeniIcerik = icerik.replace(eskiKelime,yeniKelime)
    
    with open(dosyaAdi,"w",encoding="utf-8") as file:
        file.write(yeniIcerik)

    print(f"'{eskiKelime}' kelimesi '{yeniKelime}' olarak değiştirildi.")

eskiKelime = input("Degistirilicek kelimeyi giriniz: ")
yeniKelime = input("Yeni kelimeyi girin: ")

kelimeDegistir("urunler.txt",eskiKelime,yeniKelime)

````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self, name, marka, hiz,km):
        self.name = name
        self.marka = marka
        self.hiz = hiz
        self.km = km

    def bilgileriGoster(self):
        return (f"{self.name} {self.marka} {self.hiz} {self.km}")
class Bus(Vehicle):
        def __init__(self,name,marka,hiz,km):
            self.name = name
            self.marka = marka
            self.hiz = hiz
            self.km = km

        def bilgileriGoster(self):
            return (f"Name : {self.name} Marka : {self.marka} Hiz : {self.hiz} Km: {self.km}")

bus1 = Bus("School","Volvo",180,12)
print(bus1.bilgileriGoster())

            
````
