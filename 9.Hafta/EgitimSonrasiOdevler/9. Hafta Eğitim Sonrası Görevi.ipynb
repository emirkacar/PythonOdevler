Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time 
class Kumanda:
    def __init__(self,tvDurumu="Kapali",sesDurumu=0,kanalListesi=["TRT","STAR","SHOW","KANAL D"],suankiKanal="Trt"):
        self.tvDurumu = tvDurumu
        self.sesDurumu = sesDurumu
        self.kanalListesi = kanalListesi
        self.suankiKanal=suankiKanal
        self.favoriKanalListesi = []
        self.yabanciDiller = []
        self.altYaziDurumu = 0

    def tvAc(self):
        if(self.tvDurumu=="Kapali"):
            print("Televizyon aciliyor.")
            self.tvDurumu="Acik"
        else:
            print("Televizyon zaten acik.")

    def tvKapat(self):
        if(self.tvDurumu=="Kapali"):
            print("Televizyon zaten kapali.")
        else:
            print("Televizyon kapaniyor.")
            self.tvDurumu="Kapali"
        

    def sesAyarlari(self):
        print(f"Ses seviyesi : {self.sesDurumu}")

    def sesAc(self):
        if(self.sesDurumu>=30):
            print("Ses maksimum seviyede")
        else:
            self.sesDurumu += 1
        
    def sesKapat(self):
        if(self.sesDurumu<=0):
            print("Ses minimum seviyede")
        else:
            self.sesDurumu -= 1
        

    def kanalDegistir(self):
        print(f"{self.kanalListesi}")
        rastgeleKanal=random.choice(self.kanalListesi)
        self.suankiKanal=rastgeleKanal
        print(f"Suanki kanal = {self.suankiKanal} ")
        
    
    def favoriKanalEkle(self,favoriKanal):
        self.favoriKanalListesi.append(favoriKanal)
        
    def favoriKanallariGoster(self):
        print(f"Favori kanallariniz: {self.favoriKanalListesi}")
    def altyaziyiAc(self):
        self.altYaziDurumu = 1
        print("Alt yazi aciliyor.")

    def altyaziyiKapat(self):
        self.altYaziDurumu = 0
        print("Alt yazi kapatiliyor.")

    def yabanciDilEkle(self,yabanciDil):
        self.yabanciDiller.append(yabanciDil)
        print("Yabanci dil eklendi.")
        
    def kanalEkle(self):
        kanal = input("Hangi kanali eklemek istiyorsunuz?")
        if(kanal in self.kanalListesi):
            print(f"{kanal} kanali listenizde mevcut.")
        else:
            self.kanalListesi.append(kanal)
        print(f"Kanallar:{self.kanalListesi}")

    def kanalSayisiniOgren(self):
        kanalSayisi=0
        for i in self.kanalListesi:
            kanalSayisi+=1
        print(f"Kanal sayisi:{kanalSayisi}")
        

    def rastgeleKanalaGec(self):
        print(f"Kanallar = {self.kanalListesi}")
        rastgeleKanal = random.random(self.kanalListesi)
        print(f"Rastgele kanala gecildi.Yeni kanal {rastgeleKanal}")



    def televizyonBilgileri(self):
        print(f"Televizyon acik mi:{self.tvDurumu} ses Seviyesi : {self.sesDurumu} kanal listesi: {self.kanalListesi} suanki kanal : {self.suankiKanal}")


kumanda=Kumanda(["TRT","STAR","SHOW","KANAL D"])
main=True

while(True):
    print("""
          1-Tv Ac
          2-Tv Kapat
          3-Ses Ayarlarini Ogren
          4-Kanal Ekle
          5-Kanal Sayisini Ogren
          6-Ratgele Kanala Gec
          7-Televizyon Bilgilerini Goster
          8-Menuyu tekrar Goster
          9-Ses Ac
          10-Ses Kapat
          11-Kanal Degistir
          12-Favori Kanal Ekle
          13- Alt Yaziyi Ac
          14-Alt Yaziyi Kapat
          15-Yabanci Dil Ekle
          16-Favori Kanallari Goster
          17-Cikis""")
    secim=int(input("1-8 arasinda bir secim yapiniz: "))
    main=False
    if(secim==1):
        kumanda.tvAc()
        main=True
    elif(secim==2):
        kumanda.tvKapat()
        main=True
    elif(secim==3):
        kumanda.sesAyarlari()
        main=True
    elif(secim==4):
        kumanda.kanalEkle()
        main=True
    elif(secim==5):
        kumanda.kanalSayisiniOgren()
        main=True
    elif(secim==6):
        kumanda.rastgeleKanalaGec()
        main=True
    elif(secim==7):
        kumanda.televizyonBilgileri()
        main=True
    elif(secim==8):
        main=True
    elif(secim==9):
        kumanda.sesAc()
        main=True
    elif(secim==10):
        kumanda.sesKapat()
        main=True
    elif(secim==11):
        kumanda.kanalDegistir()
        main=True
    elif(secim==12):
        favoriKanal=input("Favori kanalinizi yaziniz: ")
        kumanda.favoriKanalEkle(favoriKanal)
        main=True
    elif(secim==13):
        kumanda.altyaziyiAc()
        main=True
    elif(secim==14):
        kumanda.altyaziyiKapat()
    elif(secim==15):
        yabanciDil=input("Yabanci dil belirtiniz: ")
        kumanda.yabanciDilEkle(yabanciDil)
    elif(secim==16):
        kumanda.favoriKanallariGoster()
    elif(secim==17):
        break
    else:
        print("1-17 arasinda secim yapiniz.")
        main=True

        




    
        
```
